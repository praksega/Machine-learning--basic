```python
#creating dataset
import numpy as np
X = np.array([[1,3,5,7,9,13,20,20,21,24,26],[5,7,11,14,15,17,18,19,21,22,26]])
print(X)
X = X.T
print(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
```

```python
#normalize the data by shifting the mean to 0
X_meaned = X - np.mean(X, axis=0)
print(X_meaned)
plt.figure
plt.scatter(X_meaned[:,0],X_meaned[:,1])
plt.scatter(X[:,0],X[:,1])
```

```python
#Calculate the Co-Variance Matrix
C = np.cov(X_meaned, rowvar=False)
D = np.cov(X, rowvar=False)
print(C)
print(D)
```

```python
#Calculate Eignen values & Vector of cov matrix
eval, evec = np.linalg.eig(C)
print(eval)
print(evec)
```

```python
#sort eigen values(descending) and eigen vector
sorted_index = np.argsort(eval)[::-1]
sorted_eval = eval[sorted_index]
print(sorted_eval)
sorted_evac = evec[sorted_index]
print(sorted_evac)
```

ITS a 2d Data , so we want to represent in 1D
so n = 1

```python
n = 1
evec_subset = sorted_evac[:,0:n]
print(evec_subset)
```

```python
#Transform Data
X_reduced = np.dot(evec_subset.transpose(),X_meaned.transpose()).transpose()
print(X_reduced)
```

Direct Method for PCA

```python
from sklearn.decomposition import PCA
pca =PCA(n_components=1) #creating object of pca by telling no of dimensions i want to keep
pca.fit(X_meaned) #fitting the pca object on the data
Xr= pca.transform(X_meaned) #transforming the data based on the fitted PCA
print(Xr)
```
